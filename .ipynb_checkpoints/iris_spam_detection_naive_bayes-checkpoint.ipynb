```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.datasets import load_iris
from sklearn.feature_extraction.text import CountVectorizer
import kaggle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_path = "uciml/sms-spam-collection-dataset"

download_dir = os.path.join(os.getcwd(), "datasets/sms_spam") 
if not os.path.exists(download_dir):
    os.makedirs(download_dir)
    
    kaggle.api.dataset_download_files(dataset_path, path=download_dir, unzip=True)
```

```python
# Load the downloaded dataset CSV file
spam_data_path = os.path.join(download_dir, "spam.csv")
spam_data = pd.read_csv(spam_data_path, encoding="latin-1")
spam_data = spam_data[['v1', 'v2']]  # Select relevant columns
spam_data.columns = ['label', 'message']  # Rename columns

# Encode labels in Spam dataset (ham: 0, spam: 1)
spam_data['label'] = spam_data['label'].map({'ham': 0, 'spam': 1})
```

```python
# Visualization of SMS Spam Dataset
print("\nSMS Spam Dataset Sample:")
display(spam_data.head())
print("\nClass Distribution in SMS Spam Dataset:")
print(spam_data['label'].value_counts())
```

```python
# Plot class distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=spam_data, x='label')
plt.title("Class Distribution in SMS Spam Dataset")
plt.xlabel("Label (Ham=0, Spam=1)")
plt.ylabel("Count")
plt.show()
```

```python
# Vectorize SMS messages and display top 20 most common words
vectorizer = CountVectorizer(stop_words="english")
X_counts = vectorizer.fit_transform(spam_data['message'])
word_counts = np.array(X_counts.sum(axis=0)).flatten()
words = vectorizer.get_feature_names_out()
word_freq_df = pd.DataFrame({"word": words, "count": word_counts})

top_words = word_freq_df.nlargest(20, "count")
plt.figure(figsize=(10, 6))
sns.barplot(data=top_words, x="count", y="word")
plt.title("Top 20 Most Common Words in SMS Spam Dataset")
plt.xlabel("Frequency")
plt.show()
```

```python
# Vectorize SMS messages and display top 20 most common words
vectorizer = CountVectorizer(stop_words="english")
X_counts = vectorizer.fit_transform(spam_data['message'])
word_counts = np.array(X_counts.sum(axis=0)).flatten()
words = vectorizer.get_feature_names_out()
word_freq_df = pd.DataFrame({"word": words, "count": word_counts})


# Train-test split for SMS Spam dataset
X_spam = vectorizer.transform(spam_data['message']).toarray()
y_spam = spam_data['label'].values
X_train_spam, X_test_spam, y_train_spam, y_test_spam = train_test_split(X_spam, y_spam, test_size=0.3, random_state=42)
```

```python
# Load IRIS dataset
iris = load_iris()
X_iris = iris.data
y_iris = iris.target
```

```python
# Visualization of IRIS Dataset
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Display IRIS dataset info and class distribution
print("IRIS Dataset Sample:")
display(iris_df.head())
print("\nClass Distribution in IRIS Dataset:")
print(iris_df['target'].value_counts())
```

```python
# Pairplot to visualize feature distribution in IRIS dataset
sns.pairplot(iris_df, hue='target', palette="Set2", diag_kind="kde")
plt.suptitle("IRIS Dataset Feature Distribution by Class", y=1.02)
plt.show()
```

```python
# Split IRIS dataset
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.3, random_state=42)
```

```python
# Updated function with zero_division=0 to avoid warnings
def train_and_evaluate(model, X_train, X_test, y_train, y_test, dataset_name, results):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    results.append({'Model': model.__class__.__name__, 'Dataset': dataset_name, 'Accuracy': accuracy})
    
    # Print evaluation metrics
    print(f"Results for {model.__class__.__name__} on {dataset_name} dataset:")
    print("Accuracy:", accuracy)
    print("Classification Report:\n", classification_report(y_test, y_pred, zero_division=0))  # zero_division added
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("\n")
    
    # Plot confusion matrix
    plt.figure(figsize=(6, 4))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(f"{model.__class__.__name__} Confusion Matrix on {dataset_name}")
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.show()
```

```python
# List to hold the results for comparison
results = []
```

```python
# Apply Gaussian Naive Bayes
gnb_iris = GaussianNB()
train_and_evaluate(gnb_iris, X_train_iris, X_test_iris, y_train_iris, y_test_iris, "IRIS", results)

gnb_spam = GaussianNB()
train_and_evaluate(gnb_spam, X_train_spam, X_test_spam, y_train_spam, y_test_spam, "SMS Spam", results)
```

```python
# Apply Multinomial Naive Bayes
mnb_iris = MultinomialNB()
train_and_evaluate(mnb_iris, X_train_iris, X_test_iris, y_train_iris, y_test_iris, "IRIS", results)

mnb_spam = MultinomialNB()
train_and_evaluate(mnb_spam, X_train_spam, X_test_spam, y_train_spam, y_test_spam, "SMS Spam", results)
```

```python
# Apply Bernoulli Naive Bayes
bnb_iris = BernoulliNB()
train_and_evaluate(bnb_iris, X_train_iris, X_test_iris, y_train_iris, y_test_iris, "IRIS", results)

bnb_spam = BernoulliNB()
train_and_evaluate(bnb_spam, X_train_spam, X_test_spam, y_train_spam, y_test_spam, "SMS Spam", results)
```
